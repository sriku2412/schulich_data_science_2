```python
import numpy as np
import pandas as pd
import matplotlib as mp
import statsmodels.api as sm
```

```python
mu, sigma = 0, 5 # mean and standard deviation of normal distribution for the error term
x = np.random.uniform(40,80,100)
epsilon = np.random.normal(mu,sigma,100)
y = 3 + 4*x + epsilon
```

```python
import seaborn as sns 
sns.histplot(x,bins=20)
```

```python
model_reg = sm.OLS(y,x).fit()
model_reg.summary()
```

```python
x_updated = sm.add_constant(x)
model_updated = sm.OLS(y,x_updated).fit()
model_updated.summary()
```

```python
# We now generate autocorrelated error terms
epsilon[0] = np.random.normal(mu,sigma,1)
for i in range(0,99):
    epsilon[i+1]=0.4*epsilon[i]+0.6*np.random.normal(mu,sigma,1)
```

```python
y = 3 + 4*x + epsilon
```

```python
sns.scatterplot(x=x,y=y)
```

```python
from scipy.linalg import toeplitz
toeplitz(np.array([1,0.5,0,0,0,0,0,0]))
```

```python
rho = 0.4
cov_matrix = sigma**2*toeplitz(np.append([1,rho],np.zeros(98)))
sm.GLS(y,x_updated,cov_matrix).fit().summary()
```

```python

def mean_squared_error(true, pred):
    squared_error = np.square(true - pred) 
    sum_squared_error = np.sum(squared_error)
    mse_loss = sum_squared_error / true.size
    return mse_loss

mean_squared_error(y,predict)
```

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
method =['ward','single','complete','average','median']
for i in method:
    dendogram = sch.dendrogram(sch.linkage(points,method=i))
    plt.show()

```

```python
from sklearn.cluster import KMeans,DBSCAN
kmeans = KMeans(n_clusters=3,init='k-means++',random_state=123).fit(df)
```

```python
#plotting the clusters
plt.scatter(df['x'],df['y'],c=kmeans.labels_)
```

```python
# Applying K-means clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(df)
```

```python
# Applying DBSCAN clustering
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=15, min_samples=2)

df['dbscan_cluster'] = dbscan.fit_predict(df)

# Plotting DBSCAN scatterplot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='x', y='y', data=df, hue='dbscan_cluster', palette='Set1')
plt.title('DBSCAN Clustering')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```
